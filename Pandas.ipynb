PANDAS DATAFRAME.

```python
#creating a pandas dataframe
from sklearn.datasets import fetch_california_housing
```

Manipulating a data frame


```python
#adding a column to a dataframe
calif_df['new_column_name'] = calif_data.target
calif_df.head()
```

```python
#locating a particular column
print(calif_df.iloc[:,0])  #value of first column
print(calif_df.iloc[:,1])  #value of second column
print(calif_df.iloc[:,-1]) #values for the last column
```
