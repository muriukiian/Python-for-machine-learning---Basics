Imbalanced dataset
*   This is a dataset with an unequal class distribution.

```python
#importing the dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the csv to pamdas dataframe
credit_data = pd.read_csv('/content/credit_data.csv')
```

```python
#to view the first five rows of the dataframe
credit_data.head()
```

```python
#shape of the dataframe
credit_data.shape
```

```python
#last five rows
credit_data.tail()
```

```python
#to check for missing values
credit_data.isnull().sum()
```

```python
fig, ax = plt.subplots(figsize = (5,5))
sns.distplot(credit_data.V5)
```

```python
credit_data['V5'].fillna(credit_data['V5'].mean(), inplace = True)
```

```python
fig, ax = plt.subplots(figsize = (5,5))
sns.distplot(credit_data.V6)
```

```python
credit_data['V6'].fillna(credit_data['V6'].median(), inplace = True)
```

```python
credit_data.isnull().sum()
```

```python
#or, since only one row has missing values and the dataframe is big enough, the last row can be dropped
credit_data1 = credit_data.dropna(how = 'any')
```

```python
credit_data1.shape
```

```python
#the distribution of the two classes
credit_data1.value_counts('Class')
```

This is a highly imbalanced dataset;

*   0 ==> legitimate transactions
*   1 ==> fraudulent transactions



```python
#separating the legit and fraud transactions
legit = credit_data1[credit_data1.Class == 0]
fraud = credit_data1[credit_data1.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

==> Building a sample dataset containing similar distribution of legit and fraudulent transactions.

*   Number of fraudulent transactions is 81.


UNDER-SAMPLING

```python
legit_sample = legit.sample(n = 81)
```

```python
legit_sample.shape
```

```python
#concatenate the two dataframes
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```
